# Filtering

## 1. Creating an Artificial Signal

We'll generate a signal with a sampling rate of 100 Hz, lasting for 30 seconds. This signal will have a baseline shift and some added noise.

```python
# Import packages to this tutorial
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Parameters
sampling_rate = 100  # Hz
duration = 30  # seconds
time = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Create a simple artificial signal (e.g., a sine wave)
frequency = 5  # Hz
amplitude = 1
signal = amplitude * np.sin(2 * np.pi * frequency * time)

# Add baseline shift and noise
baseline_shift = 2
noise = np.random.normal(0, 0.5, signal.shape)
example_of_raw_signal = signal + baseline_shift + noise
```

## 2. Detrending the Signal

Now, we'll remove the baseline shift by subtracting the mean from the signal.

```python
def detrend_signal(signal):
    return signal - np.mean(signal)

detrended_signal = detrend_signal(example_of_raw_signal)
```

## 3. Building a Passband Butterworth Filter

A Butterworth filter is a type of signal processing filter designed to have as flat a frequency response as possible in the passband. It's characterized by its order and cutoff frequencies.

```python
# Functions to build the filter using scipy

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Filter parameters
order = 3
lowcut = 1.0
highcut = 10.0

# Apply the filter
filtered_signal = butter_bandpass_filter(detrended_signal, lowcut, highcut, sampling_rate, order)
```

## 4. Plotting the Results

```python
# Creating subplots
fig, axs = plt.subplots(3, 1, figsize=(14, 6))

# Plotting each signal
axs[0].plot(time, example_of_raw_signal)
axs[0].set_title('Raw Signal')
axs[0].set_ylabel('Amplitude')

axs[1].plot(time, detrended_signal)
axs[1].set_title('Detrended Signal')
axs[1].set_ylabel('Amplitude')

axs[2].plot(time[:len(filtered_signal)], filtered_signal)  # Adjust time if needed
axs[2].set_title('Filtered Signal')
axs[2].set_ylabel('Amplitude')
axs[2].set_xlabel('Time')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
